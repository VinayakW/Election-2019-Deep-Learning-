### **ENCODING THE DATA USING LABEL ENCODER.**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

df["STATE"] = le.fit_transform(df["STATE"])
df["CONSTITUENCY"] = le.fit_transform(df["CONSTITUENCY"])
df["PARTY"] = le.fit_transform(df["PARTY"])
df["GENDER"] = le.fit_transform(df["GENDER"])
df["CATEGORY"] = le.fit_transform(df["CATEGORY"])
df["EDUCATION"] = le.fit_transform(df["EDUCATION"])

df.drop("NAME",axis=1,inplace=True)

df.head()
```

###**SPLITTING THE DATA IN DEPENDENT & INDEPENDENT SET.**

```python
x = df.iloc[: , :-1].values
y = df.iloc[: , -1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y , test_size = 0.3 , random_state = 1 )
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

#**MODEL CREATION**

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
#Initialising the network
ann = Sequential()

#Adding layers to the network
ann.add(Dense(units = 12 , activation = "relu")) #1st input layer
ann.add(Dense(units = 8 , activation = "relu"))  #2nd input layer
ann.add(Dense(units = 4 , activation = "relu"))  #3rd input layer

#Creating the output layer
ann.add(Dense(units = 1 , activation = "sigmoid"))

#Establish the connection between the layers
ann.compile(optimizer = "adam" , loss = "binary_crossentropy")

#Training the model
ann.fit(xtrain,ytrain , epochs = 300 , validation_data= (xtest,ytest))
```

**Plotting the loss graph**

```python
loss = pd.DataFrame(ann.history.history)
loss.plot()
```

```python
ann.evaluate(xtrain,ytrain)
```

```python
ann.evaluate(xtest,ytest)
```

**Interpretation from graph :-**

Low training error ===> Low Bias

High testing error ===> High Variance

Hence, from above interpretation it can be said that the model is overfitting.

###**Handling Overfitting**

***Early Stop Method***

```python
ann = Sequential()
ann.add(Dense(units = 12 , activation = "relu"))
ann.add(Dense(units = 8 , activation = "relu"))
ann.add(Dense(units = 4 , activation = "relu"))

ann.add(Dense(units = 1 , activation = "sigmoid"))

ann.compile(optimizer = "adam" , loss = "binary_crossentropy" , metrics = "accuracy")

from tensorflow.keras.callbacks import EarlyStopping
earlystop = EarlyStopping(monitor = "val_loss",mode = "min", verbose = 1 , patience = 20)

ann.fit(xtrain,ytrain , epochs = 250 , validation_data= (xtest,ytest) , callbacks=[earlystop])
```

**Plotting the loss graph**

```python
losses = pd.DataFrame(ann.history.history)
losses.plot()
```

```python
ann.evaluate(xtrain,ytrain)
```

```python
ann.evaluate(xtest,ytest)
```

```python
yp = ann.predict(xtest)
```

```python
yp = yp>0.5
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest,yp))
```

####**CROSS VALIDATION**

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import KFold , cross_val_score 
```

```python
def mymodel():
  ann = Sequential()
  ann.add(Dense(units = 12 , activation = "relu"))
  ann.add(Dense(units = 8 , activation = "relu"))
  ann.add(Dense(units = 4 , activation = "relu"))
  
  ann.add(Dense(units = 1 , activation = "sigmoid"))
  
  ann.compile(optimizer = "adam" , loss = "binary_crossentropy" , metrics = "accuracy")
  
  from tensorflow.keras.callbacks import EarlyStopping
  
  earlystop = EarlyStopping(monitor = "val_loss",mode = "min", verbose = 1 , patience = 20)
  
  ann.fit(xtrain,ytrain , epochs = 250 , validation_data= (xtest,ytest) , callbacks=[earlystop])

  return ann
```

```python
estimator = KerasClassifier(build_fn= mymodel, epochs=50, batch_size=5, verbose=0 )
kf = KFold(n_splits=10, shuffle=True)
cvs = cross_val_score(estimator, x,y, cv=kf)
print(f"Avg Accuracy -: {cvs.mean()*100}\nSTD : {cvs.std()}")
```

###**The average accuracy of model is 82% approx.**
